Este website teve design e programação desenvolvidos por Rafael Badain com programação adicional por Irina Lerner. A interface gráfica foi produzida com a biblioteca Bootstrap e os códigos podem ser encontrados em nosso GitHub. Ícones por Becris, Inipagi Studio, KonKapp e Just Icon.
- github: https://github.com/covid19br/covid19br.github.io
- contato: obscovid19br@gmail.com

#### Fontes Principais
Utilizamos os registros das notificações de casos de COVID-19 divulgadas diariamente pelo Ministério da Saúde. A cada atualização na plataforma um programa automático de recuperação de dados nos envia os novos registros e dispara a atualização das análise e gráficos neste site.

#### Outras Fontes:

- Portal sobre Coronavírus da FIOCRUZ: https://portal.fiocruz.br/coronavirus
- Jonh Hopkins Coronavirus Resource Center: https://coronavirus.jhu.edu/map.html
- Site sobre epidemia de COVID-19 da Organização Mundial da Saúde: https://www.who.int/health-topics/coronavirus#tab=tab_1
- CMMID COVID repository, Centre for Mathematical Modelling of Infectious Diseases, London School of Hygiene and Tropical Medicine.: https://cmmid.github.io/topics/covid19/

#### Importante:
- Projeções de número de casos

Com o modelo linear ajustado estimamos o número de casos para os próximos dias, e os intervalos de confiança dos valores previstos

- Análise de erro das nossas previsões

Fazendo previsões do número de casos confirmados para os próximos 5 dias

- Série temporal dos tempos de duplicação

Para avaliar o quanto o crescimento da epidemia varia ao longo do tempo, ajustamos o modelo linear para períodos de cinco dias

# Outros sites:

- https://labs.wesleycota.com/sarscov2/br/ – Criado originalmente pelo doutorando Wesley Cota, envolve a colaboração da Universidade Federal de Viçosa e a Universidade de Zaragoza na Espanha. Mostra a variação dos casos para o Brasil, por estado e por município.

- https://sites.google.com/view/covidcids/p%C3%A1gina-inicial?authuser=0 – Página mantida por um grupo da Faculdade de Medicina da Universidade do Porto, com informações e análises do COVID 19 em Portugal.

- https://cmmid.github.io/topics/covid19/ – Um excelente portal de conteúdo da London School of Hygiene & Tropical Medicine. Contém várias análises e disponibiliza ferramentas para análise e predição da situação do COVID 19 (em inglês).

- https://who.maps.arcgis.com/apps/webappviewer/index.html?id=2203b04c3a5f486685a15482a0d97a87&extent=-18853893.9193%2C-508973.7482%2C-2103789.289%2C8511818.5819%2C102100 – Site da OPAS/OMS apresentando os casos de coronavírus para as Américas. O site mostra os casos que apareceram em cada país das Américas, rastreando a origem da infecção.

- https://sigageomarketing.com.br/coronavirus/ – Desenvolvido pela startup paranaense SIGA, o sistema mostra online os casos atualizados de Covid-19 por estado, no Brasil.

- https://coronavirus.jhu.edu/map.html – O Centro de Ciência e Engenharia de Sistemas da Universidade Johns Hopkins desenvolveu um mapa interativo que atualiza os casos do novo coronavírus (covid-2019) ao redor do mundo em tempo real (em inglês).

- https://nssac.bii.virginia.edu/covid-19/dashboard/ – Apresenta um Painel de vigilância do COVID-19 para as Américas e para o mundo (em inglês).
- https://datastudio.google.com/u/0/reporting/f6ad0988-f203-45f8-8d18-5d726c1d2d8b/page/MGzDB – Apresenta uma mapa de Casos e Clusters, dia a dia, do COVID-19 fora da China (em inglês).

- https://www.healthmap.org/covid-19/ – Apresenta um mapa animado da evolução dos casos no mundo desde o começo da epidemia (em inglês).

- https://covid19-brazil-api.now.sh/ – Disponibiliza uma API com dados sobre os casos de doença COVID-19 no Brasil e mundo.

- https://milkeninstitute.org/covid-19-tracker – O Instituto Milken está acompanhando o desenvolvimento de tratamentos e vacinas para o COVID-19 (coronavírus). Este documento contém uma agregação de informações publicamente disponíveis de fontes validadas

- https://www.transparimed.org/single-post/2020/03/27/COVID-19-clinical-trials-information-sources – Mais de 500 ensaios clínicos de potenciais tratamentos e intervenções com COVID-19 foram lançados em todo o mundo

- https://covid-evidence.org/ – é um banco de dados atualizado continuamente das evidências disponíveis em todo o mundo sobre intervenções para o COVID-19
- http://hbiostat.org/proj/covid19/ – Resources for COVID-19 Randomized Clinical Trial Design Vanderbilt Institute for Clinical and Translational Science Research Methods Program Department of Biostatistics Vanderbilt University School of Medicine Nashville TN USA

- https://www.medrxiv.org/content/10.1101/2020.03.24.20041020v1 – Systematic review and critical appraisal of prediction models for diagnosis and prognosis of COVID-19 infection

- https://www.comitecientifico-ne.com.br – Comitê Científico do Consórcio Nordeste

- https://www.covid19.ifnoar.com.br/indicadores – desenvolvido por pesquisadores do Instituto Federal de Goiás e colaboradores, visa apresentar informações (número de casos, óbitos, taxa de contaminação,…) sobre o vírus COVID-19 pelo Brasil.

- https://www.instagram.com/covidcafepod/?hl=pt-br – COVIDCafé. Um podcast que discute aspectos de ciência e epidemiologia relevantes para a pandemia da COVID-19 através de linguagem acessível.

```python
xM = dfConfObitoMunCopy.Município
yC = dfConfObitoMunCopy.Confirmados
yO = dfConfObitoMunCopy.Óbitos
yP = dfConfObitoMunCopy.População
```

```python
mun = "PICOS"
conf = int(dfConfObitoMunCopy[dfConfObitoMunCopy.Município == mun].Confirmados)
ob = int(dfConfObitoMunCopy[dfConfObitoMunCopy.Município == mun].Óbitos)
po = int(dfConfObitoMunCopy[dfConfObitoMunCopy.Município == mun].População)
labels = "Confirmados" , "Óbitos", "População"
sizes = conf,ob,po
explode = (0.0,0.0,1.2)

plt.pie(sizes,labels=labels,explode=explode,autopct="%1.1f%%",shadow=False,startangle=0)
plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(1,1,figsize=(15,40))

y_pos = np.arange(len(xM))
yClist =  list(yC)

ax.barh(y_pos, yClist, align='center',label="a")
ax.set_yticks(y_pos)
ax.set_yticklabels(xM)
ax.invert_yaxis() 
ax.set_xlabel('Quantida de casos')
plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots(1,1,figsize=(15,40))

y_pos = np.arange(len(xM))
yOlist =  list(yO)

ax.barh(y_pos, yOlist, align='center',label="a",color="red")
ax.set_yticks(y_pos)
ax.set_yticklabels(xM)
ax.invert_yaxis() 
ax.set_xlabel('Quantida de casos')
plt.show()
```
